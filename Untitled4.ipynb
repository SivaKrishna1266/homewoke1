```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
# The downloads are included in the source file but suppressed here for cleaner output
# nltk.download('punkt', quiet=True)
# nltk.download('stopwords', quiet=True)
# nltk.download('averaged_perceptron_tagger', quiet=True)
# nltk.download('wordnet', quiet=True)
nltk.download('punkt_tab') # Added download for punkt_tab

text = "John enjoys playing football while Mary loves reading books in the library."
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to map NLTK POS tags to WordNet tags for accurate lemmatization
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# 1. Segment into tokens [cite: 135]
tokens = word_tokenize(text)
print("Tokens:", tokens)

# 2. Remove stopwords [cite: 136]
# Also filters out punctuation (e.g., the final period '.') with word.isalpha()
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("\nAfter removing stopwords:", filtered_tokens)

# 3. Apply lemmatization (requires POS tagging) [cite: 137]
pos_tags = pos_tag(filtered_tokens)
lemmas = [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in pos_tags]
print("\nLemmatized words:", lemmas)

# 4. Keep only verbs and nouns [cite: 138]
verbs_nouns = [lemmatizer.lemmatize(word, get_wordnet_pos(pos))
               for word, pos in pos_tags if pos.startswith('N') or pos.startswith('V')]
print("\nOnly Verbs and Nouns:", verbs_nouns)
```

```python
# ==========================
# Q2 â€“ NER + Pronoun Ambiguity Detection
# ==========================

# Import necessary NLTK modules
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

# --- Download all required NLTK resources ---
# These ensure all models and tokenizers are available
nltk.download('punkt', quiet=True)                    # sentence & word tokenizer
nltk.download('punkt_tab', quiet=True)                # updated tokenizer data (for newer NLTK versions)
nltk.download('maxent_ne_chunker', quiet=True)        # main named entity chunker model
nltk.download('maxent_ne_chunker_tab', quiet=True)    # updated chunker model data (newer NLTK)
nltk.download('words', quiet=True)                    # English word list (needed for NE recognition)
nltk.download('averaged_perceptron_tagger', quiet=True)       # POS tagger (older name)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)   # POS tagger (new name in latest versions)

# --- Input text to analyze ---
text = "Chris met Alex at Apple headquarters in California. He told him about the new iPhone launch."

# --- Step 1: Tokenize and POS-tag the text ---
tokens = word_tokenize(text)     # splits text into individual words (tokens)
pos_tags = pos_tag(tokens)       # assigns each token a part-of-speech label (NN, VB, etc.)

# --- Step 2: Perform Named Entity Recognition (NER) ---
# ne_chunk groups tagged words into named entities (like PERSON, ORGANIZATION, GPE)
chunks = ne_chunk(pos_tags)

print("Named Entities:")
for chunk in chunks:
    if hasattr(chunk, 'label'):  # only process items that are named-entity subtrees
        entity = " ".join(c[0] for c in chunk)   # extract the actual words from the chunk
        print(f"{entity} --> {chunk.label()}")   # print entity name and its label type

# --- Step 3: Check for pronoun ambiguity ---
# Define a set of pronouns that often cause reference confusion
pronouns = {"he", "she", "they"}

# If any of these pronouns are present in the text, print a warning message
if any(word.lower() in pronouns for word in tokens):
    print("\nWarning: Possible pronoun ambiguity detected!")
```
